# Analyseur de sentiment : VADER

Import des mots vides du lexique anglais

```python
nltk.download('stopwords')
stopwords = stopwords.words('english')
analyzer = SentimentIntensityAnalyzer()
```

```python
df = df[df['tweet'].notnull()]
```

```python
df['clean_tweet'] = df['tweet'].apply(lambda x: ' '.join([item for item in x.split() if item not in stopwords]))
```

On va calculer le score de sentiment de chaque lignes

*   -1 : Sentiment n√©gatif
*   0  : Sentiment neutre
*   1  : Sentiment positif

Dans le cadre d'une classfication binaire, nous avons decid√© de ne pas retenir le sentiment neutre, pour avoir seulement 2 √©tiquettes



```python
df['vader_score'] = df['clean_tweet'].apply(lambda x: analyzer.polarity_scores(x)['compound'])

df.sample(100)
```

```python
positive_num = len(df[df['vader_score'] >=0.01])
negative_num = len(df[df['vader_score']<0.01])

df['vader_sentiment_label']= df['vader_score'].map(lambda x:int(1) if x>=0.01 else int(0))
```

On compte le nombre d'√©tiquette pour chaque classe

```python
nbr_label = df['vader_sentiment_label'].value_counts()
print(nbr_label)
```
