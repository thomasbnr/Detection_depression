# Import des jeux de donn√©es

La fonction read_csv() prend un certain nombre de param√®tres :


*   delimiter = "," les valeurs de chaque ligne sont s√©par√©s par des virgules
*   engine = 'python', le moteur de parsing est Python
*   on_bad_lines = 'skip', les lignes incorrectes sont ignor√©es plut√¥t que de g√©n√©rer une erreur.
*   encoding="utf-8", l'encodage utilis√© est l'utf-8
*   header=[0], la premi√®re ligne contient les noms des colonnes
*   index_col[0], on pr√©cise que la premi√®re colonne du fichier doit √™tre sp√©cifi√©e comme colonne d'index pour le dateFrame



```python
depressed = pnd.read_csv('depressed.csv',delimiter=',',engine='python',on_bad_lines="skip",encoding="utf-8", header=[0],index_col=[0])
depressed.head()
```

```python
depression = pnd.read_csv('depression.csv',delimiter=',',engine='python',on_bad_lines="skip",encoding="utf-8", header=[0],index_col=[0])
depression.head()
```

```python
hopeless = pnd.read_csv('hopeless.csv',delimiter=',',engine='python',on_bad_lines="skip",encoding="utf-8", header=[0],index_col=[0])
hopeless.head()
```

```python
lonely = pnd.read_csv('lonely.csv',delimiter=',',engine='python',on_bad_lines="skip",encoding="utf-8", header=[0],index_col=[0])
lonely.head()
```

```python
suicide = pnd.read_csv('suicide.csv',delimiter=',',engine='python',on_bad_lines="skip",encoding="utf-8", header=[0],index_col=[0])
suicide.head()
```

# Nettoyage de notre jeu de donn√©es

Suppression des colonnes inutiles

```python
new_df = pnd.concat([depression, hopeless, lonely, depressed, suicide], ignore_index=True)
new_df.drop(['time','date', 'timezone', 'username', 'name', 'conversation_id', 'created_at', 'user_id', 'place', 'likes_count', 'hashtags','cashtags','link', 'retweet', 'quote_url', 'video', 'user_rt_id', 'near', 'geo', 'mentions', 'urls', 'photos', 'replies_count', 'retweets_count'], axis = 1, inplace = True)
new_df.head()
```

Apr√®s la fusion de nos datasets, notre jeu de donn√©es contient les lignes en fonction des th√®mes. Dans l'ordre nous avons :

*   depressed
*   depression
*   hopeless
*   lonely
*   suicide

Nous devons donc m√©langer nos lignes.


```python
depressed_df = new_df.sample(frac=1).reset_index(drop=True)
```

```python
depressed_df = depressed_df.head(100000)
```

On remarque que certaines lignes ne sont pas en anglais, il faut donc les supprimer

```python
def detect_language(text):
    try:
        return detect(text)
    except:
        return ''
```

```python
depressed_df ['language'] = depressed_df ['tweet'].apply(detect_language)

depressed_df.head()
```

```python
# Filtrage des lignes en anglais
filtered_lines = depressed_df [depressed_df ['language'] == 'en']

filtered_lines = filtered_lines.reset_index(drop=True)
```

```python
final_df = filtered_lines
final_df.tail()
```

On exporte notre jeu de donn√©es final

```python
export_csv = final_df.to_csv(r'depressive_tweet.csv')

# T√©l√©charger le fichier CSV
files.download('depressive_tweet.csv')
```
